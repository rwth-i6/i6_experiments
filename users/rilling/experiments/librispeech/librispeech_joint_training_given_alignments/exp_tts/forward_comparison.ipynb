```python
globs = [
    "/u/lukas.rilling/experiments/glow_tts_asr_v2/alias/experiments/librispeech/joint_training/given_alignments/raw_audio/TTS_models/glowTTS*/enc768/200ep/dec_drop_0.05/tts_eval_gl/forward",
    "/u/lukas.rilling/experiments/glow_tts_asr_v2/alias/experiments/librispeech/joint_training/given_alignments/raw_audio/TTS_models/glowTTS_simple_encoder/*/200ep/dec_drop_*/tts_eval_gl/forward",
]
lr_files = []
for g in globs:
    lr_files += glob.glob(g)

common_prefix = os.path.commonpath(lr_files)
common_sufix = os.path.commonpath([l[::-1] for l in lr_files])[::-1]

names = []
for f in lr_files:
    names.append(f.removeprefix(common_prefix).removesuffix(common_sufix))

files = dict(zip(lr_files, names))

sequence_names = os.listdir(list(files.keys())[0] + "/output/audio_files/")

files
```

```python
sequence_index = 100
sequence_name = sequence_names[sequence_index]

for folder, name in files.items():
    path = folder + "/output/audio_files/" + sequence_name
    print(path)
    if os.path.exists(path):
        with open(folder.replace("forward", "nisqa_mos/average").replace("alias", "output")) as f:
            nisqa_result = f.readline()
        print(name)
        print("AutoMOS: " + nisqa_result)
        display(Audio(path))
```
